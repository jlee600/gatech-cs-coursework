## Q1: KMeans Clustering [40pts + 3pts Grad / 1.5% Bonus for Undergrad]

KMeans is trying to solve the "hard clustering problem." Given N data points, $X$, and an integer K, the objective is to find a disjoint cover of the dataset into K clusters, $S = \{S_1\subseteq X, S_2\subseteq X, \ldots, S_K\subseteq X\}$, where every data point belongs to exactly 1 of the clusters. We also need to pick an associated "center/centroid" for each cluster. The objective, then, is as follows:

$$ \arg \min_S \sum_{i=1}^K \sum_{\forall x \in S_i} \|x - \mu_i\|^2 $$

We need to find the optimal clustering and optimal centers such that the sum of the distances from each point to their assigned cluster is minimized. The choice of $\mu_i$ is forced by this definition. If we know a clustering is correct, then the optimal center for each cluster is the center of mass of the assigned points (hence why the "centers" are also called "centroids").

KMeans solves this with an iterative process very similar to EM. Start with some guess for the centers, assign each point to the closest center, then set the centers to the center of mass of the points assigned to it.

### 1.1 Pairwise Distance [5pts]

For that problem, we need to calculate the distances from a set of points (data) to another set of points (centers). If we want to use a lot of data, this function has to be fast and space efficient! We'll implement a function that calculates all distances at once into a "pairwise distance matrix."

Given $X \in \mathbb{R}^{N \times D}$ and $Y \in \mathbb{R}^{M \times D}$, obtain the pairwise Euclidean distance matrix $\textsf{dist} \in \mathbb{R}^{N \times M}$, where $\textsf{dist}_{i,j} = \|X^{\{i\}} - Y^{\{j\}}\|_2 = \sqrt{\sum_{d=1}^D \left(X^{\{i\}}_d - Y^{\{j\}}_d\right)^2}$.

Any naive looping will be too slow, and if we use the broadcasting trick for pairwise Manhattan distance from homework 1, we'll have an $\mathbb{R}^{N\times M\times D}$ intermediate array, which might be too big to allocate in memory. You can, of course, assume that $\mathbb{R}^{N\times M}$ will fit in memory. If it didn't, we would need to segment our return to some kind of stream. But how do we do this without looping or the dimension expansion broadcasting trick?

Consider a single element in the pairwise distance matrix: (i,j). We'll expand that pair's distance, and hopefully get a process that can be applied at the same time for every pair.

$$\textsf{dist}_{i,j} = \sqrt{\sum_{d=1}^D \left(X^{\{i\}}_d - Y^{\{j\}}_d\right)^2}$$

Expand out the squared term,

$$\textsf{dist}_{i,j} = \sqrt{\sum_{d=1}^D \left[{X^{\{i\}}_d}^2 + {Y^{\{j\}}_d}^2 - 2 X^{\{i\}}_d Y^{\{j\}}_d \right]}$$

Then recollect like terms into their own summation,

$$\textsf{dist}_{i,j} = \sqrt{\sum_{d=1}^D \left[{X^{\{i\}}_d}^2\right] + \sum_{d=1}^D \left[{Y^{\{j\}}_d}^2\right] - \sum_{d=1}^D \left[2 X^{\{i\}}_d Y^{\{j\}}_d \right]}$$

You already know how to get a array containing the row sums of the squared elements for both X and Y, you did that in the programming warmup for HW1. Doing this, you'll get results shaped like (N,) and (M,), respectively. If we want to solve for $\textsf{dist}_{i,j}$, we just index into the X results with i and index into the Y results with j.

However, we want this process to be able to output a pairwise matrix shaped like (N,M). The X term is the same for any row of Y we choose, so we could shape the X results like (N,1) and then repeat it M times to get (N,M). The Y term is the same for any row of X we choose, so we could shape the Y results like (1,M) and repeat it N times to get (N,M). Fortunately, broadcasting will do this for you. If you use the `keepdims` argument (or np.newaxis) and a careful transpose to get your squared row sum arrays shaped like (N,1) and (1,M), when you add those two arrays together, numpy will automatically repeat each array along the empty axis to get an expanded result (N,M), where result[i,j] will be $\sum_{d=1}^D \left[{X^{\{i\}}_d}^2\right] + \sum_{d=1}^D \left[{Y^{\{j\}}_d}^2\right]$. That's exactly what we want! [(R, 1) and (R,) are not the same shape](https://stackoverflow.com/questions/22053050/difference-between-numpy-array-shape-r-1-and-r), and this summation trick shows why we might want to keep the empty axes around.

Now, all that's left is that cross term, which is actually a dot product!

$$\textsf{dist}_{i,j} = \sqrt{\sum_{d=1}^D \left[{X^{\{i\}}_d}^2\right] + \sum_{d=1}^D \left[{Y^{\{j\}}_d}^2\right] - \sum_{d=1}^D \left[2 X^{\{i\}}_d Y^{\{j\}}_d \right]}$$
$$\textsf{dist}_{i,j} = \sqrt{\sum_{d=1}^D \left[{X^{\{i\}}_d}^2\right] + \sum_{d=1}^D \left[{Y^{\{j\}}_d}^2\right] - 2\left(\sum_{d=1}^D \left[X^{\{i\}}_d Y^{\{j\}}_d \right]\right)}$$
$$\textsf{dist}_{i,j} = \sqrt{\sum_{d=1}^D \left[{X^{\{i\}}_d}^2\right] + \sum_{d=1}^D \left[{Y^{\{j\}}_d}^2\right] - 2\left(X^{\{i\}} \cdot Y^{\{j\}} \right)}$$

Again, if we were just solving for $\textsf{dist}_{i,j}$, we just get the $i^{th}$ row from X, $j^{th}$ row from Y, and do the dot product. However, we want to be able to output a pairwise matrix for all of the results.

Luckily, matrix multiplication calculates the pairwise dot product between the rows of the first argument and the columns of the second argument. We want the pairwise dot product between the rows of X and the rows of Y, so you'll need to flip the rows into columns for one of the arguments. As a hint, look at the shapes. You have X as (N,D) and Y as (M,D) and you want a result that's (N,M). How do you organize the terms so a matrix multiplication gives you (N,M)?

Your implementation needs to use this technique. Looping is not permitted. Making an intermediate array that is larger than the output array is not permitted.

### 1.3 Clustering Metrics [8pts]

In this section, you will calculate metrics to assess the quality of your clustering algorithm. 

1. The **Adjusted Rand Index (ARI)** quantifies how consistently two clusterings group pairs of points, correcting for chance.
2. The **Silhouette Coefficient** is a measure of how well a data point fits within its assigned cluster compared to other clusters.

#### 1.3.1 Adjusted Rand Index (ARI) [3pts]

As discussed in class, the computation for the Random Index is as follows:

$$RI = \frac{TP + TN}{TP + TN + FP + FN}$$

Using this value, we can calculate the Adjusted Random Index score, which is simply the random index score adjusted for chance (since cluster labels are not some derived thing, you can flip labels and the clustering is functionally the same).
$$ARI = \frac{RI - \mathbb{E}[\text{RI}]}{max(RI) - \mathbb{E}[\text{RI}]}$$
where
$$ max(RI) = 1 $$
$$\mathbb{E}[\text{RI}] = \frac{(TP+FP)(TP+FN) + (TN+FP)(TN+FN)}{(TP + TN + FP + FN)^2} $$
Substituting these values into the $ARI$ formula, we see that
$$ARI = \frac{2(TP \cdot TN - FP \cdot FN)}{(TP + FN)(FN + TN) + (TP + FP)(FP + TN)}$$


TP (True Positive) represents the number of data points that are correctly clustered together in both the algorithm's result and the ground truth.

TN (True Negative) is the number of data points that are correctly placed in separate clusters in both the algorithm's result and the ground truth.

FP (False Positive) counts the number of data points that are incorrectly clustered together in the algorithm's result but correctly separated in the ground truth.

FN (False Negative) is the number of data points that are incorrectly separated in the algorithm's result but correctly clustered together in the ground truth.

#### 1.3.2 Silhouette Coefficient [5pts]

As discussed in class, the computation for the Silhouette Coefficient is as follows:

$$s(i) = \frac{b(i) - a(i)}{\max\{a(i), \, b(i)\}}$$

where

$a(i)$ = average distance of point i to all other points in its own cluster

$b(i)$ = minimum average distance of point i to all points in any other cluster

Thus, the overall silhouette coefficient can be computed as the average of the silhouette coefficient for all points:

$$ S = \frac{1}{N} \sum_{i=1}^{N} s(i) $$

Now for your task: you will need to pick what you believe is the optimal value of k for KMeans clustering.

There is a ground-truth that is technically correct. This data is derived from a dense embedding of images in a classification problem. There were a discrete number of classes, which you could use for k, but we're not telling. Frankly, the ground-truth doesn't always matter in terms of clustering. Even though there were x number of classes in the original problem, that might not be reflected in the dense embedding. Also, on most problems, you're not gonna know. The k-selection problem is genuinely an issue in clustering, because "How many clusters are there?" is a terribly vague question. You're not going to be graded based on the correctness of your choice of k, but rather on how well your evidence supports your conclusion.

Qualitative analysis of the scatter plot is valid, since "How many clusters are there?" is a qualitative question. ### 2.2 Background Material [0pts]

A Gaussian Mixture Model (GMM) is a probabilistic model that assumes all the data points are generated from a mixture of a finite number of Gaussian distributions called components. GMM can be thought of as a soft clustering algorithm, since modelling the underlying distribution as a set of mixed components means we can derive how much each point "belongs" to each component, and we can just call our components "clusters."

Given $N$ samples $X = [x^{\{1\}}, x^{\{2\}}, \ldots, x^{\{N\}}]$, where $x^{\{i\}} \in \mathbb{R}^D$. Let $\pi_k$ be the mixing coefficient indicating the prevalence of the $k^{th}$ Gaussian component, let $\mu_k$ be the mean ($\mu_k \in \mathbb{R}^D$) of the $k^{th}$ Gaussian component, and let $\Sigma_k$ be the covariance matrix ($\Sigma_k \in \mathbb{R}^{D\times D}$) of the $k^{th}$ Gaussian component. There are $K$ total components.

The objective of GMM is to model the underlying distribution of a set of data, that is, we want to build the probability density function that best represents the data. You can think of this as the pdf that is most likely to have made our data set. The space of all possible pdfs is impossibly huge, so we limit ourselves to the following form:

$$f(x) = \sum_{k=1}^K \pi_k \cdot \mathcal{N}(x; \mu_k, \Sigma_k)$$

Try plotting out this function with 1D Gaussians and 2 components. You'll see that the $\pi_k$ essentially just scales down the Gaussian, and adding them together just superimposes them. The result of this expression is a bunch of lumps. The objective, then, is to place the lumps wherever there is a cluster of data, choose the covariance matrix to spread out the lump enough, and scale the Gaussian by the proportion of the data contained in that lump. This is why GMM is successful as a clustering algorithm, since "finding a model that accurately represents the underlying distribution $P(X)$ of the data" is tantamount to locating the clusters.

More accurately, we need to fit the parameters $\mu_k$, $\Sigma_k$, and $\pi_k$ for each component $k$. We'll use the iterative EM formula that you may have already seen a glimpse of in the written question.

From lecture, you may already know about the tool we'll use: responsibility, $\tau$. Each point is assigned a responsibility $\tau$ to each component based on how well that component's mean and covariance represent the datapoint (the likelihood of the datapoint under the Gaussian).

- **E-step:** Evaluate the responsibility matrix $\tau$.

$$
\tau_{nk}=P\left(z=k\Big\vert x^{\{n\}}\right)=\frac{\pi_{k} \cal{N}\left(x^{\{n\}} | \mu_{k}, \Sigma_{k}\right)}{\sum_{j=1}^{K} \pi_{j} \cal{N}\left(x^{\{n\}} | \mu_{j}, \Sigma_{j}\right)}, \quad \text{for } k = 1, \dots, K
$$

We let the responsibility of a point $x^{\{n\}}$ to a component $k$ to be the likelihood of $x^{\{n\}}$ under that component divided by the sum of the likelihoods under every component. Note that every data point has a probability for each component. Thus, $\tau$ has a size $N\times K$, which we can keep in a single matrix. You'll need to be clever about how you compute this, but once you get the likelihoods for each point under each component, getting responsibilities is just dividing each value by its row's sum.

- **M-step:** Re-estimate Paramaters

Using the responsibilities, we can do MLE on a tight lower bound of the total likelihood to get the best estimate relative to those responsibilities. Essentially $\tau$ acts as a latent variable. If we knew the true responsibilities of each point, we could jump straight to the answer, but since we don't, we need to iteratively refine our estimates.

You should use the following update rules:

$$ N_k = \sum_{n=1}^N \tau_{nk} $$

$$ \mu_k^{new} = \frac{1}{N_k} \cdot\sum_{n=1}^N \tau_{nk} x^{\{n\}} $$

$$ \Sigma_k^{new} = \frac{1}{N_k} \cdot\sum_{n=1}^N \tau_{nk}(x^{\{n\}} - \mu_k^{new})(x^{\{n\}}-\mu_k^{new})^T $$

$$ \pi_k^{new} = \frac{N_k}{N} $$
Note that the vector multiplication term on the covariance calculations should be an outer product that results in a $D\times D$ matrix, not an inner product that results in a scalar.

- We repeat E and M steps until the incremental improvement to the likelihood function is small.

#### 2.6.3. Resampling [8pts]

Our next objective is to create artificial data that resembles the initial distribtion. There are several methods for sampling from a pdf, but we'll use a technique called "rejection sampling." Essentially, if we uniformly fill up the 3d space that our wireframe graph was plotted on, then threw away all points above the graph, we would keep more points where the graph is taller. This process is statistically indistinguishable from true generation.

To be more precise:
1. Pick a candidate $c\in\mathbb{R}^d$ uniformly at random from the support of the pdf (all regions of non-zero probability density).
2. Compute the probability density of that point under the model, $p(c)$; for us, of course, $p(c) = \sum_{k=1}^{K} \pi_k \cdot \mathcal{N}(c ;\; \mu_k, \Sigma_k)$
3. Use this probability density as the probability of acceptance. To actualize this:
     - Pick a uniform random number between 0.0 and 1.0, $u\sim U[0,1]$ (the uniform distribution).
     - If $u \leq p(c)$, accept $c$, else, reject $c$.

This [source from Cosmic Coding](https://cosmiccoding.com.au/tutorials/rejection_sampling) is a pretty good tutorial (for 1D rejection sampling) if you'd like to see another explainer.

This process, though, is really bad for us because of how slow it will be, especially for the GMM. Unfortunately, Gaussians have infinitely long tails, which means the support of our GMM is the entire real space $\mathbb{R}^d$. This means that the candidates can be any number in the infinite space. For one, that makes step 1 impossible. There is no way to define a uniform pdf over an infinite domain (you would need $\int_{-\infty}^\infty\int_{-\infty}^\infty c\; dxdy = 1$). We need to accept some reductions to make this work for us. We'll use the following procedure to hopefully finish before the heat death of the universe:
1. Pick a candidate $c\in\mathbb{R}^2$ uniformly at random inside a bounded domain. Pick $c=[x\;y]$ with $x\sim U[\textsf{xmin}, \textsf{xmax}]$ and $y\sim U[\textsf{ymin}, \textsf{ymax}]$.
2. Compute the probability density of that point under the model, $p(c) = \sum_{k=1}^{K} \pi_k \cdot \mathcal{N}(c ;\; \mu_k, \Sigma_k)$.
3. Find the maximum probability density of that the model achieves $\textsf{dmax} \leftarrow \max(p)$
4. Pick a uniform random number between $0.0$ and $\textsf{dmax}$, $u\sim U[0,\textsf{dmax}]$.
5. If $u \leq p(c)$, return $c$.
6. Else, try again.

This revised process is mostly accurate, except for the sharp bounding. Under a Gaussian, a true generator really might spit out any point in $\mathbb{R}^2$, but our sampler is sharply limited by xmin, xmax, ymin, ymax. This still might be too slow. If we want to speed up the process, we could parallelize the candidate picking, or we could even lower the threshold dmax (which would bias our sampler against picking from the modes that are snipped off). However, this should be fast enough for our purposes.

#### 3.1.2. k-Nearest Neighbors [2.5% Bonus For All]

This data has missing values that are "Missing Completely At Random" (MCAR), that is, the missingness is not correlated with the data. If the missingness were correlated with observed values, it would be "Missing At Random" (MAR), or if the missingness were correlated with the missing values themselves, it would be "Missing Not At Random" (MNAR). Luckily, the MCAR case is pretty easy to work with, and simple approaches give great results.

We'll use the k-Nearest Neighbors algorithm, which imputes missing values by taking the mean of that value from the k-neighborhood of a point (the k points closest to it). This algorithm really doesn't have much to it other than that. Compute pairwise distance, find the k smallest distances (excluding yourself obviously), take the mean, then slot it in.

The major complication here is actually the distance metric. We can't use our good old Euclidean distance because there are nan's in our data. What's $(\textsf{nan}-4.2)^2$? What's $(\textsf{nan}-\textsf{nan})^2$? We'll use a very simple analogue; when computing the Euclidean distance between x and y, if either x or y are missing in a given feature, simply ignore that term. Then, apply a correction factor: $\sqrt{\frac{D}{D_{valid}}}$, where $D_{valid}$ is the number of features where neither x nor y are missing.

Here's an example, which will hopefully be more illustrative.

$$X = [1.0, \textsf{nan}, 3.0, 4.0, \textsf{nan}]$$
$$Y = [2.0, 2.0, \textsf{nan}, 1.0, \textsf{nan}]$$

These vectors have 5 features, so $D=5$.

Feature 0 is "valid", Feature 1 is "invalid" since $X$ has a nan, Feature 2 is "invalid" since $Y$ has a nan, Feature 3 is "valid", and Feature 4 is "invalid" since both have a nan. This means that $D_{valid}=2$.

Now we do Euclidean distance, ignoring the invalid terms:

$$\textsf{dist}'=\sqrt{(1.0-2.0)^2 + (\textsf{nan}-2.0)^2 + (3.0-\textsf{nan})^2 + (4.0-1.0)^2 + (\textsf{nan}-\textsf{nan})^2}$$
$$\textsf{dist}'=\sqrt{(-1)^2 + 0 + 0 + (3)^2 + 0}$$
$$\textsf{dist}'=\sqrt{10}$$

Then, we apply the correction factor:

$$\textsf{dist} = \sqrt{\frac{D}{D_{valid}}} \cdot \textsf{dist}'$$
$$\textsf{dist} = \sqrt{\frac{5}{2}} \cdot \sqrt{10}$$
$$\textsf{dist} = \sqrt{25} = 5.0$$

## Q4: Hierarchical Clustering [15pts]

### 4.1 Implementation [15pts]

Hierarchical Clustering is a bottom-up agglomerative clustering algorithm which iteratively combines the closest pair of clusters. Each point starts off as its own cluster, and in each iteration you'll find the closest clusters and update the distances to the new cluster using single-link clustering, keeping track of the order in which the clusters are combined.

In `hierarchical_clustering.py`, you'll implement **create_distance_matrix**, **iterate**, and **fit**.

The `HierarchicalClustering` class has several instance variables that you may need to create and update in each iteration:
1. `points`: (N, D) numpy array where N is the number of points, and D is the dimensionality of each point. This is your dataset.
2. `distance`: (N, N) symmetric numpy array which stores pairwise distances between clusters. The distance between a cluster and itself should be `np.inf` in order to help us calculate the closest pair later
3. `cluster_ids`: (N,) numpy array where index_array[i] gives the cluster id of the i-th column and i-th row of distances. Initially, each point with index `points[i, :]` is assigned cluster id i, and new points are assigned cluster ids starting from `N` and incrementing.
4. `clustering`: (N - 1, 4) numpy array that keeps track of which clusters were merged in each iteration. `clustering[iteration_number]` keeps track of the first cluster id, second cluster id, distance between first and second clusters, and the size of new cluster
5. `cluster_sizes` (2N - 1, ) numpy array that stores the number of points in each cluster, indexed by id. Because there are `N` original clusters corresponding to each point, and each iteration merges two clusters, there will be `2N-1` total clusters created.

Theses are the following functions you'll have to implement in **hierarchical_clustering.py**:

1. `create_distances`: Creates the initial distance matrix and cluster ids
2. `iterate`: Merges the two closest clusters
3. `fit`: Calls `iterate` multiple times and returns the clusterings
